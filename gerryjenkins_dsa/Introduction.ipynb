#### Abstract Data Type (ADT)
- Similar to abstraction of code using procedure abstraction, but for data
- Create an abstraction for new data types
- The Client of an ADT has a set interface they can rely on certain predefined operations to work

The whole point is that an ADT provides an **implemtation independent** view of data

#### Date ADT Example 
An example of an interface method for an ADT is a date
1) get the number days to another date
2) get the week number of the year
3) advance the date by a number of days
4) set the date to m/d/y
5) return the date in a specific format
        
        Feb 2, 2014 - 2/2/14 - February 2nd, 2014
